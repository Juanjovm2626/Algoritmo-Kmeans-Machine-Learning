##Implementación del K-means con distancia Euclidiana.

**FUNCION DISTANCIA EUCLIDIANA**

```python
import random #Libreria para generar numeros aleatorios

#definir la veriable euclidiana
def distacia_euclidiana(x,c):
  return np.sqrt(np.sum((x-c)**2))
```

**Codigo kmeans sin usar librerias**

```python

#paso 1: Numero de clusters
k=4

#paso 2: eleccion de centroides
dimension= X_.shape[0]#numero de puntos dimension es el tamaño de filas que tiene dataframe
centroids=X_[random.sample(range(dimension),k)]
centroids

#almacenamiento etiquetas
clusters_manuales = np.zeros(dimension)

#almacenamietno de distancias
distancias = np. zeros ((dimension,k))


#paso 3: Implementacion del metodo y tolerancia
tol=1e-6
error=100

#repetir hasta que los centroides dejen de moverse significativamente
while error > tol:
  #Asiganacion a los clusters mas cercanos
  for i in range(dimension):
    for j in range(k):
      distancias[i,j]= distacia_euclidiana(X_[i],centroids[j])
    clusters_manuales[i]=np.argmin(distancias[i])
  #almacenar los centroides previos antes de actualizarlos nuevamente
  centroids_prev = np.copy(centroids)
  #actualizar centroides
  for l in range(k):
    puntos_cluster = X_[clusters_manuales==l]
    #si el cluster no esta vacio, recalcular el centroide
    if len(puntos_cluster)>0:
      centroids[l]= np.mean(puntos_cluster, axis=0)
    else:
      #si el cluster esta vacio, dejar el centroide
      print(f'cluster{l}vacio, reasignando centride aleatoreamente')
      centroids[l] = X_[np.random.choice(dimension)]
  #calcular el error como el cambio de promedio en los centroides
  error = np.mean([distacia_euclidiana(centroids[l], centroids_prev[l]) for l in range(k)])
  (print(error))
```
