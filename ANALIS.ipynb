**10 LIGAS CON MAYOR CANTIDAD DE JUGADORES**

```python
##Extraemos las ligas con value_counts para los valores y head para solo extraer las 10
ligas=datos["League"].value_counts().head(10)

#grafica
plt.figure(figsize=(6,4))#tamaño de la grafica
ligas.plot(kind="barh",  color= 'black')#variable, bar si la quiero vertical-barh si la quiero horizontal
plt.title('Ligas con mayor cantidad de jugadores')#titulo
plt.xlabel('Ligas')#eje x
plt.ylabel('Cantidad de jugadores')#eje y
plt.show()#mostrar
```

**Distribucion del "Overall"**

```python
bins_=np.arange(40,100,1)#arreglo desde 40 hasta 100 de 1 en 1
#crear histograma
x=datos["OVR"] #Se hace con los datos de OVR
plt.figure(figsize=(10,5)) #Tamaño de la grafica
sns.histplot(x,bins= bins_,kde=True, color= 'c')#Kde muestra la linea entre barras de histograma
plt.title('distribucion del Overall') #Titulo
plt.show()
```

**Overall por posicion**


```python
posiciones=datos.groupby('Position')['OVR'].median().sort_values().index
# median para sacar la mediana, sort_values orgnizarlo de manera ascendente y index para mostrar solo los encabezados
posiciones #agrupar position con OVR
```

```python
#diagrama de cajas (boxplot)
import warnings
warnings.filterwarnings('ignore') #ignora los warnings

#crear diagrama de cajas
attributes = ['PAC', 'SHO', 'DRI', 'PHY', 'DEF', 'PAS'] #HAcemos con los atributos
sns.boxplot(x='Position', #nombre de la variable den el dataframe, eje x
            y='OVR', #nombre de la bariable dependiente
            data=datos, #conjunto de datos, donde saca informacion
            order= posiciones, #variable creada anteriormente para indice
            palette= 'Accent') #colores

plt.title('Overall por posiciones')
```

```python
import warnings
warnings.filterwarnings('ignore')
attributes = ['PAC', 'SHO', 'DRI', 'PHY', 'DEF', 'PAS'] ##Atributos necesarios

# Configurar la figura y los ejes para los subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(18, 15), sharey=True,)
fig.suptitle('Player attributes by position', fontsize=16)

#Se ajustan los ejes para faciitar la iteracion
axes = axes.flatten()

# Graficar la distribución de cada atributo en un subplot separado
for i, attr in enumerate(attributes):
    order = datos.groupby('Position')[attr].median().sort_values().index

    # Graficar el diagrama de caja (boxplot) con categorías ordenadas
    sns.boxplot(x='Position', y=attr, data=datos, ax=axes[i], order=order, palette = 'PiYG')
    axes[i].set_title(f'{attr} rating by position')
    axes[i].set_ylabel(f'{attr}')
    axes[i].set_xlabel('Position')

# se ajusta el diseño para evitar superposicion
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
